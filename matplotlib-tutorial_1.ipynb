```python
import matplotlib.pyplot as plt
import numpy as np
```

<h1> Simple plot </h1>

```python
x_axis = np.linspace(-np.pi, np.pi, 300)
sin = np.sin(x_axis)
cos = np.cos(x_axis)
tan = np.tan(x_axis)



plt.xlim(-np.pi, np.pi)
xlabels = ['-pi','-halfpi','0','halfpi','pi']

plt.subplot(1, 3, 1)
plt.plot(x_axis, sin, color="blue", linewidth=1, linestyle="-", label="sine")
plt.xticks(np.linspace(-np.pi, np.pi, 5),xlabels)
plt.yticks([-1,0,1])
plt.title('Sine curve')
plt.ylabel('yvalue')
t = np.pi/2
plt.scatter([t, ], [np.sin(t), ], 50, color='blue')
plt.legend(loc='upper left')
plt.annotate('peak',
             xy=(t, np.sin(t)), xycoords='data',
             xytext=(+30, +60), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.3"))


plt.subplot(1, 3, 2)
plt.ylim(-1,1)
plt.plot(x_axis, cos, color="green", linewidth=2, linestyle="--", label="cosine")
plt.xticks(np.linspace(-np.pi, np.pi, 5),xlabels)
plt.yticks([-1,0,1])
plt.title('Cosine curve')
plt.legend(loc='upper right')



plt.subplot(1, 3, 3)
plt.ylim(-3,3)
plt.plot(x_axis, tan, color="red", linewidth=3, linestyle="-.", label="tangent")
plt.xticks(np.linspace(-np.pi, np.pi, 5),xlabels)
plt.yticks([-1,0,1])
plt.title('Tangent curve')
plt.legend(loc='lower center')



plt.show()
    
```
